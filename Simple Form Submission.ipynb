```python
import tkinter as tk
from tkinter import messagebox

class FormApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Simple Form Submission")
        self.root.geometry("400x400")

        tk.Label(root, text="Name:", font=('Helvetica', 12)).pack(anchor='w', padx=20, pady=5)
        self.name_entry = tk.Entry(root, width=30)
        self.name_entry.pack(padx=20)

        tk.Label(root, text="Email:", font=('Helvetica', 12)).pack(anchor='w', padx=20, pady=5)
        self.email_entry = tk.Entry(root, width=30)
        self.email_entry.pack(padx=20)

        tk.Label(root, text="Gender:", font=('Helvetica', 12)).pack(anchor='w', padx=20, pady=5)
        self.gender_var = tk.StringVar()
        tk.Radiobutton(root, text="Male", variable=self.gender_var, value="Male").pack(anchor='w', padx=40)
        tk.Radiobutton(root, text="Female", variable=self.gender_var, value="Female").pack(anchor='w', padx=40)

        tk.Label(root, text="Skills:", font=('Helvetica', 12)).pack(anchor='w', padx=20, pady=5)
        self.skill_python = tk.IntVar()
        self.skill_java = tk.IntVar()
        self.skill_js = tk.IntVar()
        tk.Checkbutton(root, text="Python", variable=self.skill_python).pack(anchor='w', padx=40)
        tk.Checkbutton(root, text="Java", variable=self.skill_java).pack(anchor='w', padx=40)
        tk.Checkbutton(root, text="JavaScript", variable=self.skill_js).pack(anchor='w', padx=40)

        tk.Button(root, text="Submit", command=self.submit_form, bg='green', fg='white').pack(pady=10)
        tk.Button(root, text="Reset", command=self.reset_form).pack()

    def submit_form(self):
        name = self.name_entry.get().strip()
        email = self.email_entry.get().strip()
        gender = self.gender_var.get()
        skills = []

        if self.skill_python.get(): skills.append("Python")
        if self.skill_java.get(): skills.append("Java")
        if self.skill_js.get(): skills.append("JavaScript")

        if not name or not email or not gender:
            messagebox.showwarning("Input Error", "Please fill in all required fields.")
            return

        info = f"Name: {name}\nEmail: {email}\nGender: {gender}\nSkills: {', '.join(skills)}"
        messagebox.showinfo("Submitted Info", info)

        # Optional: save to file
        with open("form_data.txt", "a") as f:
            f.write(info + "\n" + "-"*40 + "\n")

    def reset_form(self):
        self.name_entry.delete(0, tk.END)
        self.email_entry.delete(0, tk.END)
        self.gender_var.set("")
        self.skill_python.set(0)
        self.skill_java.set(0)
        self.skill_js.set(0)

if __name__ == "__main__":
    root = tk.Tk()
    app = FormApp(root)
    root.mainloop()
```
